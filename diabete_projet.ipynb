```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Pour affichage plus lisible
pd.set_option('display.max_columns', None)

# Charger les données (ex: depuis un fichier CSV)
df = pd.read_csv('diabetes_dataset.csv')  # nom du fichier ou chemin réel
```

```python
# Premières lignes
print(df.head())

# Dimensions
print(df.shape)

# Types et valeurs manquantes
print(df.info())

# Statistiques descriptives
print(df.describe())

# Nombre de valeurs uniques par colonne
print(df.nunique())
```

```python
# Vérifier les zéros dans des colonnes qui ne devraient pas en avoir
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in cols_with_zero:
    print(f'{col}: {(df[col] == 0).sum()} zéros')

# Remplacer les 0 par NaN pour les colonnes concernées
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)

# Remplir les NaN avec la médiane (ou moyenne, ou KNN imputation etc.)
df.fillna(df.median(), inplace=True)
```

```python
# Histogrammes
df.hist(bins=30, figsize=(15,10))
plt.tight_layout()
plt.show()

# Densité par Outcome
for col in df.columns[:-1]:
    sns.kdeplot(data=df, x=col, hue='Outcome', shade=True)
    plt.title(f'Distribution de {col} selon Outcome')
    plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de corrélation')
plt.show()
```

```python
sns.pairplot(df, hue='Outcome')
plt.show()
```
